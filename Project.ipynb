```python
import pandas as pd
# Open and reshape.
data = pd.read_csv('datasets/stackoverflow/stackOverFlow.csv', index_col = 0)
data = data[['Country', 'Student', 'Employment', 'FormalEducation', 'UndergradMajor', 'CompanySize', 'DevType', 'YearsCoding', 'JobSatisfaction', 'CareerSatisfaction', 'HopeFiveYears', 'JobSearchStatus', 'LastNewJob', 'SalaryType', 'ConvertedSalary', 'LanguageWorkedWith', 'OperatingSystem', 'NumberMonitors', 'StackOverflowRecommend', 'StackOverflowVisit', 'HoursComputer', 'HoursOutside', 'Age']]
# print(data.shape[0])
# print(data.shape[1])
# data
```

```python
# Chages string to boolean.
data['Student'] = data['Student'].replace(['Yes, part-time', 'Yes, full-time'], True)
data['Student'] = data['Student'].replace('No', False)
# data
```

```python
# Gets data info.
counter = 0
data['NumberMonitors'] = data['NumberMonitors'].replace('More than 4', '5')
data['NumberMonitors'] = data['NumberMonitors'].astype(float)
data['StackOverflowRecommend'] = data['StackOverflowRecommend'].replace('10 (Very Likely)','10')
data['StackOverflowRecommend'] = data['StackOverflowRecommend'].replace('0 (Not Likely)','0')
data['StackOverflowRecommend'] = data['StackOverflowRecommend'].astype(float)
# for i in data:
#     print('Name and tag in column ' + data.columns[counter] + ': ')
#     print('Number of empty in column ' + data.columns[counter] + ': ' + str(data[i].isnull().sum()))
#     print('Type in column ' + data.columns[counter] + ': ' + str(data[i].dtypes))
#     if(str(data[i].dtypes) == 'float64'):
#         print('Type of data in column ' + data.columns[counter] + ': ' + 'numeric, float64')
#     else:
#         if((data[i][1] == True) or (data[i][1] == False)):
#             print('Type of data in column ' + data.columns[counter] + ': ' + 'binary, bool')
#         else:
#             print('Type of data in column ' + data.columns[counter] + ': ' + 'other, string, etc')
#     print('\n---------------------------------------------------------------------------\n')
#     counter += 1
```

```python
# The amount of salary for everyone on full-time job.
normalsalary = []
for index, row in data.iterrows():
    if(str(row[2]) == 'Employed part-time'):
        normalsalary.append(2 * row[14])
    elif(str(row[2]) == 'Employed full-time'):
        normalsalary.append(row[14])
    else:
        normalsalary.append(data['ConvertedSalary'].mean())
for i in range(0, len(normalsalary)):
    if(str(normalsalary[i]) == 'nan'):
        normalsalary[i] = data['ConvertedSalary'].mean()
for i in range(0, len(normalsalary)):
    normalsalary[i] = int(int(normalsalary[i]))
# print(normalsalary)
```

```python
data.insert(15, 'Normal Salary', normalsalary, True)
# data
```

```python
# Chages interval to category.
data['CompanySize'] = data['CompanySize'].replace('Fewer than 10 employees', 'Enthusiasts')
data['CompanySize'] = data['CompanySize'].replace('10 to 19 employees', 'Medium Group')
data['CompanySize'] = data['CompanySize'].replace('20 to 99 employees', 'Small Company')
data['CompanySize'] = data['CompanySize'].replace('100 to 499 employees', 'Standart Company')
data['CompanySize'] = data['CompanySize'].replace('500 to 999 employees', 'Large Company')
data['CompanySize'] = data['CompanySize'].replace('1,000 to 4,999 employees', 'Minor Corporation')
data['CompanySize'] = data['CompanySize'].replace('5,000 to 9,999 employees', 'Average Corporation')
data['CompanySize'] = data['CompanySize'].replace('10,000 or more employees', 'Global Corporation')
# data
```

```python
# Dependence of being student on worktime schedule.


# 0 means totaly free person.
# 1 means no work student.
# 2 means part-time work adult.
# 3 means full-time work adult.
# 4 means part-time work student.
# 5 means full-time work student.


datanew = data[['Student', 'Employment']]
workstudy = []
for index, row in datanew.iterrows():
    if(row[0]):
        # Student.
        if(row[1] == 'Employed full-time'):
            workstudy.append(5)
        elif(row[1] == 'Employed part-time'):
            workstudy.append(4)
        else:
            workstudy.append(1)
            
    else:
        # Not student.
        if(row[1] == 'Employed full-time'):
            workstudy.append(3)
        elif(row[1] == 'Employed part-time'):
            workstudy.append(2)
        else:
            workstudy.append(0)


workstudy.sort()
# plt.xlabel('Types of worktime schedule')
# plt.ylabel('Number of people')
# plt.title('Dependence of being student on worktime schedule')
a = [0, 0, 0, 0, 0, 0]
for i in range(0, 6):
    a[i] = workstudy.count(i)
x = [i for i in range(0, len(a))]
# plt.bar(x, height = a, width=0.8, color='g')
```

```python
# Windows is most popular OS for multiple screens.


oses = dict()
oses['win'] = [0, 0, 0, 0]
oses['mac'] = [0, 0, 0, 0]
oses['lin'] = [0, 0, 0, 0]
oses['unb'] = [0, 0, 0, 0]
for index, row in data.iterrows():
    if(row[17] == 'Windows'):
        try:
            oses['win'][int(row[18]) - 1] = oses['win'][int(row[18]) - 1] + 1
        except:
            continue
    elif(row[17] == 'MacOS'):
        try:
            oses['mac'][int(row[18]) - 1] = oses['mac'][int(row[18]) - 1] + 1
        except:
            continue
    elif(row[17] == 'Linux-based'):
        try:
            oses['lin'][int(row[18]) - 1] = oses['lin'][int(row[18]) - 1] + 1
        except:
            continue
    elif(row[17] == 'BSD/Unix'):
        try:
            oses['unb'][int(row[18]) - 1] = oses['unb'][int(row[18]) - 1] + 1
        except:
            continue


# print(oses)
```

```python
datamons = data.pivot_table('NumberMonitors', index=['CompanySize','Student'], columns = 'OperatingSystem')
for i in datamons.columns:
    datamons[i] = datamons[i].map(lambda x: round(x, 2))
datamons = datamons.reindex([('Enthusiasts',False),('Enthusiasts',True),
                            ('Medium Group',False),('Medium Group',True),
                            ('Small Company',False),('Small Company',True),
                            ('Standart Company',False),('Standart Company',True),
                            ('Large Company',False),('Large Company',True),
                            ('Minor Corporation',False),('Minor Corporation',True),
                            ('Average Corporation',False),('Average Corporation',True),
                            ('Global Corporation',False),('Global Corporation',True)])
datamons.to_excel('datamons.xlsx')
```
